```python
# read by default 1st sheet of an excel file
summary_stats = pd.read_excel('../results/astro_VE_vs_H3DEGene_statistics.xlsx')
```

```python
pos_significant_genes = summary_stats[(summary_stats['padj'] < 0.05) & (summary_stats['log2FoldChange'] > 0.25)]
neg_significant_genes = summary_stats[(summary_stats['padj'] < 0.05) & (summary_stats['log2FoldChange'] < -0.25)]
```

```python
pos_significant_genes
```

```python
# Gene lists for enrichment analysis
pos_gene_list = list(pos_significant_genes['gene'].astype(str))
neg_gene_list = list(neg_significant_genes['gene'].astype(str))
```

```python
neg_results = gp.enrichr(gene_list=neg_gene_list, organism='Mouse', gene_sets='GO_Biological_Process_2023',
                             outdir=None, cutoff=0.05)
```

```python
neg_results.res2d.Term = neg_results.res2d.Term.str.split(" \(GO").str[0]
```
